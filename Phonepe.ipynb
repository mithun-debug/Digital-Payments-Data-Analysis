```python
datasets = {
    "State Txn & Users": state_txn_users,
    "State Txn Split": state_txn_split,
    "State Device Data": state_device_data,
    "District Txn & Users": dist_txn_users,
    "District Demographics": dist_demographics
}

for name, df in datasets.items():
    print(f"--- Analysis for: {name} ---")
    print("\n1. Data Types:")
    print(df.info())
    print("\n2. Summary Statistics:")
    print(df.describe())
    print("-" * 30)
```

```python
#1.3 finding missing values
for name, df in datasets.items():
    print(f"--- Missing Values in {name} ---")
    
    null_counts = df.isnull().sum()
    null_percentages = (null_counts / len(df)) * 100
    
    # Filter only columns that actually have missing values
    missing_info = null_percentages[null_percentages > 0]
    
    if not missing_info.empty:
        print(missing_info)
        # Highlight the highest percentage
        highest_col = missing_info.idxmax()
        print(f"Highest Missing Value: {highest_col} ({missing_info[highest_col]:.2f}%)")
    else:
        print("No missing values found!")
    print("\n")
```

```python
#1.4 
total_states = dist_txn_users['State'].nunique()
total_districts = dist_txn_users['District'].nunique()

state_max_districts = dist_txn_users.groupby('State')['District'].nunique().idxmax()
max_count = dist_txn_users.groupby('State')['District'].nunique().max()

print(f"Total States: {total_states}")
print(f"Total Districts: {total_districts}")
print(f"State with most districts: {state_max_districts} ({max_count} districts)")
```

```python
#2.6 Line Plot
selected_state = 'Karnataka'
state_data = state_txn_users[state_txn_users['State'] == selected_state]

time_labels = state_data['Year'].astype(str) + "-Q" + state_data['Quarter'].astype(str)

plt.figure(figsize=(12, 6))

plt.plot(time_labels, state_data['App Opens'], marker='o', linestyle='-', color='orange', linewidth=2)

plt.title(f'App Usage Trend in {selected_state}', fontsize=14)
plt.xlabel('Timeline (Year-Quarter)', fontsize=12)
plt.ylabel('Number of App Opens', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#2.7 Distribution of Transaction Types
latest_year = state_txn_split['Year'].max()
latest_quarter = state_txn_split[state_txn_split['Year'] == latest_year]['Quarter'].max()

recent_data = state_txn_split[(state_txn_split['Year'] == latest_year) & 
                               (state_txn_split['Quarter'] == latest_quarter)]

type_distribution = recent_data.groupby('Transaction Type')['Transactions'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
type_distribution.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title(f'Distribution of Transaction Types ({latest_year} - Q{latest_quarter})', fontsize=14)
plt.xlabel('Transaction Type', fontsize=12)
plt.ylabel('Total Transactions', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# 3.1 Data Consistency Across Levels
district_totals = dist_txn_users.groupby(['State', 'Year', 'Quarter']).agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum',
    'Registered Users': 'sum'
})
```

```python
district_totals
```

```python
state_totals = state_txn_users.set_index(['State', 'Year', 'Quarter'])[['Transactions', 'Amount (INR)', 'Registered Users']]
discrepancies = district_totals - state_totals
mismatches = discrepancies[(discrepancies != 0).any(axis=1)]

if mismatches.empty:
    print("Success: Data is consistent across State and District levels!")
else:
    print("Discrepancies found between District-level and State-level data:")
    print(mismatches)
```

```python
#4.1 Data merging
# 1. Sum population at the state level from District Demographics
state_pop = dist_demographics.groupby('State')[['Population']].sum()

latest_users = state_txn_users.groupby('State')[['Registered Users']].max()

adoption_analysis = latest_users.join(state_pop)

adoption_analysis['User_to_Pop_Ratio'] = adoption_analysis['Registered Users'] / adoption_analysis['Population']

print("User-to-Population Ratio per State:")
print(adoption_analysis[['User_to_Pop_Ratio']])
```

```python
import matplotlib.pyplot as plt

adoption_sorted = adoption_analysis.sort_values(by='User_to_Pop_Ratio', ascending=False)

plt.figure(figsize=(16, 8))
plt.bar(adoption_sorted.index, adoption_sorted['User_to_Pop_Ratio'], color='mediumpurple')

plt.xlabel('State', fontsize=12)
plt.ylabel('User to Population Ratio', fontsize=12)
plt.title('Digital Adoption Rate: Registered Users vs. Total Population', fontsize=15)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
#4.2 Correlation between population density and transaction
district_merged = pd.merge(dist_txn_users, dist_demographics, on=['State', 'District'])

correlation = district_merged['Density'].corr(district_merged['Transactions'])
print(f"Correlation between Population Density and Transaction Volume: {correlation:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(district_merged['Density'], district_merged['Transactions'], alpha=0.5, color='coral')

plt.title('Population Density vs. Transaction Volume', fontsize=14)
plt.xlabel('Density (People per sq km)', fontsize=12)
plt.ylabel('Total Transactions', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# 4.3 Average transaction amount per user
user_spend_data = latest_users.join(state_atv_data[['Amount (INR)']])
user_spend_data['Avg_Spent_Per_User'] = user_spend_data['Amount (INR)'] / user_spend_data['Registered Users']

top_5_user_spend = user_spend_data.sort_values(by='Avg_Spent_Per_User', ascending=False).head(5)
bottom_5_user_spend = user_spend_data.sort_values(by='Avg_Spent_Per_User', ascending=False).tail(5)

print("Top 5 States - Highest Average Spent per User:")
print(top_5_user_spend[['Avg_Spent_Per_User']])

print("\nBottom 5 States - Lowest Average Spent per User:")
print(bottom_5_user_spend[['Avg_Spent_Per_User']])
```

```python
#4.4 Device brand usage ratio
device_ratio_df = pd.merge(state_device_data, state_txn_users[['State', 'Year', 'Quarter', 'Registered Users']], 
                           on=['State', 'Year', 'Quarter'], 
                           suffixes=('_Brand', '_Total'))

device_ratio_df['Usage_Ratio'] = device_ratio_df['Registered Users_Brand'] / device_ratio_df['Registered Users_Total']

latest_device_stats = device_ratio_df[device_ratio_df['Year'] == latest_year]
print("Device Brand Usage Ratios (Latest Year):")
print(latest_device_stats[['State', 'Brand', 'Usage_Ratio']].head(10))
```

```python
target_state = 'Maharashtra'
state_devices = latest_device_stats[latest_device_stats['State'] == target_state].sort_values(by='Usage_Ratio', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(state_devices['Brand'], state_devices['Usage_Ratio'], color='crimson')

plt.title(f'Device Brand Usage Ratio in {target_state}', fontsize=14)
plt.xlabel('Mobile Brand', fontsize=12)
plt.ylabel('Usage Ratio', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#5.1 Total Transactions and Total Amount over time for a selected state
import matplotlib.pyplot as plt

# 1. Filter data for the selected state
state_name = 'Karnataka'
state_df = state_txn_users[state_txn_users['State'] == state_name]
time_axis = state_df['Year'].astype(str) + "-Q" + state_df['Quarter'].astype(str)

# 2. Create the plot with dual axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Transactions on the left axis
ax1.plot(time_axis, state_df['Transactions'], color='blue', marker='o', label='Transactions')
ax1.set_ylabel('Total Transactions', color='blue')

# Create a second axis for the Amount
ax2 = ax1.twinx()
ax2.plot(time_axis, state_df['Amount (INR)'], color='green', marker='s', label='Amount (INR)')
ax2.set_ylabel('Total Amount (INR)', color='green')

plt.title(f'Transaction Trends over Time for {state_name}')
plt.xticks(rotation=45)
plt.show()
```
